#GARCH (chạy thử, không viết báo cáo)


```r
install.packages("fGarch")
```

```r
library(fGarch)
```

```r
install.packages('rugarch', dependencies = TRUE)
```

```r
library(rugarch)
```

```r
arima_model <- auto.arima(gold2)
```

```r
arima_model
```

```r
residuals <- residuals(arima_model)
residuals
```

```r
# Khớp mô hình GARCH với phần dư
garch_model <- ugarchspec(variance.model = list(model = "sGARCH",
                                                garchOrder = c(1, 1)),
                          mean.model = list(armaOrder = c(1, 1),
                                            include.mean = TRUE),
                          distribution.model = "std")
fit <- ugarchfit(garch_model, residuals)
```

```r
print(fit)
```

```r
forecasts = ugarchforecast(fit, n.ahead = 15)
```

```r
forecasts
```
